```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
df = pd.read_excel("my_transactions table.xlsx")
df
```

```python
df.columns
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
#Filling the  missing values of the following variables 
fill_values = ["online_order","brand", "product_line", "product_class", "product_size" ]
for label, content in df.items():
    for i in fill_values:
        if label == i:
            if pd.isnull(content).sum():
                df[label] = content.fillna(method = "ffill")
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
#fill in the float missing values 
for label, content in df.items():
    if pd.api.types.is_float_dtype(content):
        if pd.isnull(content).sum():
            df[label] = content.fillna(content.median())
```

```python
df.isna().sum()
```

```python
#Add the profit column to the transactions dataset
df["profit"] = df["list_price"] - df["standard_cost"]
```
